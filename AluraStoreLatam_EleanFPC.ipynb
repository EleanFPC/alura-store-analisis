### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Cálculo de ingresos por tienda (precio + costo de envío)

ingresos_tienda1 = tienda["Precio"].sum() + tienda["Costo de envío"].sum()
ingresos_tienda2 = tienda2["Precio"].sum() + tienda2["Costo de envío"].sum()
ingresos_tienda3 = tienda3["Precio"].sum() + tienda3["Costo de envío"].sum()
ingresos_tienda4 = tienda4["Precio"].sum() + tienda4["Costo de envío"].sum()

print("Ingresos tienda 1:", ingresos_tienda1)
print("Ingresos tienda 2:", ingresos_tienda2)
print("Ingresos tienda 3:", ingresos_tienda3)
print("Ingresos tienda 4:", ingresos_tienda4)
```

En este análisis calculamos los ingresos totales de cada tienda sumando el precio de venta y su costo de envío. Esto nos permite identificar qué tienda genera más ingresos y cuál tiene un desempeño más débil.


# 2. Ventas por categoría

```python
import matplotlib.pyplot as plt

categorias1 = tienda["Categoría del Producto"].value_counts()

plt.figure(figsize=(12,6))
plt.bar(categorias1.index, categorias1.values)
plt.title("Ventas por categoría - Tienda 1")
plt.xlabel("Categoría")
plt.ylabel("Cantidad de productos vendidos")
plt.xticks(rotation=45)
plt.show()

categorias1
```

Este gráfico muestra la distribución de ventas por categoría en esta tienda. Nos ayuda a identificar cuáles categorías son las más relevantes comercialmente y cuáles presentan menor rotación.

# 3. Calificación promedio de la tienda


```python
print("Calificación promedio por tienda:")
print("Tienda 1:", tienda["Calificación"].mean())
print("Tienda 2:", tienda2["Calificación"].mean())
print("Tienda 3:", tienda3["Calificación"].mean())
print("Tienda 4:", tienda4["Calificación"].mean())
```

La calificación promedio indica el nivel de satisfacción de los clientes con cada tienda. Las tiendas con menor calificación pueden estar teniendo problemas de servicio o producto.


# 4. Productos más y menos vendidos

```python
mas_vendidos1 = tienda["Producto"].value_counts().head(5)
menos_vendidos1 = tienda["Producto"].value_counts().tail(5)

print("Productos más vendidos - Tienda 1:")
print(mas_vendidos1)

print("\nProductos menos vendidos - Tienda 1:")
print(menos_vendidos1)
```

Este análisis permite visualizar qué productos tienen mayor rotación y cuáles presentan un desempeño más bajo. Esto ayuda en decisiones de inventario y liquidación.

# 5. Envío promedio por tienda

```python
print("Envío promedio:")

print("Tienda 1:", tienda["Costo de envío"].mean())
print("Tienda 2:", tienda2["Costo de envío"].mean())
print("Tienda 3:", tienda3["Costo de envío"].mean())
print("Tienda 4:", tienda4["Costo de envío"].mean())
```

El costo de envío promedio permite evaluar la eficiencia logística de cada tienda. Un envío más costoso o inconsistente afecta la experiencia del cliente y los ingresos.
